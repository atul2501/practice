```python
batsman=pd.read_csv('/content/sharma-kohli.csv')
batsman

```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#labels titles
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('performace of rohit and virat')
plt.xlabel('season')
plt.ylabel('scores')
```

```python
#colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue')
plt.title('performace of rohit and virat')
plt.xlabel('season')
plt.ylabel('scores')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dashed',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth=3)
plt.title('performace of rohit and virat')
plt.xlabel('season')
plt.ylabel('scores')
#dashdot,dotted,dash-datted,solid
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dashed',linewidth=3,marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth=3,marker='+')
plt.title('performace of rohit and virat')
plt.xlabel('season')
plt.ylabel('scores')
#marker=d,+ and many type
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dashed',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth=3,marker='+',markersize=10)
plt.title('performace of rohit and virat')
plt.xlabel('season')
plt.ylabel('scores')

#markersize joha points bante hai vo
```

```python
#legend -->location
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dashed',linewidth=3,marker='D',markersize=10,label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth=3,marker='+',markersize=10,label='rohit')
plt.title('performace of rohit and virat')
plt.xlabel('season')
plt.ylabel('scores')

plt.legend(loc='upper left')
#loc choose automatic its best
```

```python
#limiting axes
price = [48000,54000,49000,45000,47000,4500000]
year= [2015,2016,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(1,10000000)
plt.xlim(2015,2020)
```

# Pie Chart




1.   Univariate/bivariate analysis
2.   Categorical vs numerical
1.   Use case - To find contibution on s standard scale





```python
#simple data
data=[23,45,100,20,49]
suject=['eng','science','maths','sst','hindi']

plt.pie(data,labels=suject)
plt.show()
```

```python
df=pd.read_csv('/gayle-175.csv')
```

```python
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','red','yellow','orange','pink'])
plt.show()
```

```python
#explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode =[0,0.1,0,0.3,0.1,0],shadow=True)
plt.show()
```
